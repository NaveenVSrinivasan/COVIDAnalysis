```python
from IPython.core.display import display, Javascript, HTML
import json
```

```python
def display_vis(file_name):
    display(HTML(filename="vis.html"))
    with open("build_vis.js","r") as vis_file: 
        script_code = vis_file.read()
    command = script_code + "\ngenerate_vis({},element)".format(file_name)
    jqWrapper = "(function(element){" + command +  "})(element);"
    display(Javascript(jqWrapper))
    
```

```python
display_vis("'./EMBEDDINGS_PROCESSED'")
```
